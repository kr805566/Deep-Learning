```python
import numpy as np
```

```python
Arr = np.random.randint(0,10,size=[12,12])
```

```python
print(Arr)
```

```python
ps = 2
N = int(12/ps)
Ans=[]
for i in range(N):
  An=[]
  for j in range(N):
    Arr1 = Arr[i*ps:i*ps+ps,j*ps:j*ps+ps]
    An.append(np.max(Arr1))
  Ans.append(An)

np.around(Ans, 2) 
```
