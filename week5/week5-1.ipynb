```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.set_printoptions(linewidth=1000)
```

```python
y = np.linspace(0.01, 0.99, 1000)
yh = np.linspace(0.01, 0.99, 1000)
Y, YH = np.meshgrid(y,yh)
ZZ = -(Y*np.log(YH)+(1-Y)*np.log(1-YH))

fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(1,2,1,projection='3d')
ax1.plot_surface(Y, YH, ZZ, cmap='jet')
ax1.set_xlabel('y', fontsize=14)
ax1.set_ylabel('$\hat{y}$', fontsize=14)


ax2 = fig.add_subplot(1,2,2)

cs = ax2.contour(Y, YH, ZZ,[0.3,0.5,0.6,0.69315,1.0,1.2,1.8,4.0])
ax2.clabel(cs, inline=1, fontsize=8, fmt='%.5f')
ax2.set_aspect('equal')
ax2.set_xlabel('y', fontsize=14)
ax2.set_ylabel('$\hat{y}$', fontsize=14)


plt.show()
```
