```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
def f(x, y):
  return (1-x)**2 + 100*(y-x**2)**2
def derivativeX(x, y, h=0.001):
  return (f(x+h,y)-f(x-h,y))/(2*h)
def derivativeY(x, y, h=0.001):
  return (f(x,y+h)-f(x,y-h))/(2*h)
```

```python
x = np.linspace(-2, 2, 1000)
y = np.linspace(-1, 3, 1000)

XX, YY = np.meshgrid(x,y)

fig = plt.figure(figsize=(18,6))

ax1 = fig.add_subplot(1,2,1,projection='3d')
ax1.plot_surface(XX, YY, f(XX,YY), cmap=plt.cm.jet)
ax1.set_xlabel('x', fontsize=14)
ax1.set_ylabel('y', fontsize=14)



ax2 = fig.add_subplot(1,2,2)

conf = ax2.contourf(XX, YY, f(XX, YY), 8, alpha=0.75, cmap=plt.cm.jet)
cs = ax2.contour(XX, YY, f(XX,YY))
ax2.clabel(cs, inline=1, fontsize=8, fmt='%.5f')
ax2.set_aspect('equal')



plt.show()
```

```python
def run(w0, w1):
  aX = []
  aY = []

  eta = 0.04/30
  beta = 0.968/30

  v0 = w0*eta
  v1 = w1*eta
  wp0, vp0 = w0, v0
  wp1, vp1 = w1, v1
  aX.append(w0)
  aY.append(w1)
  for i in range(1000):
    wp0, vp0 = w0, v0
    wp1, vp1 = w1, v1
    v0 = beta*vp0 + eta*derivativeX(w0, w1)
    v1 = beta*vp1 + eta*derivativeY(w0, w1)

    w0 = wp0-v0
    w1 = wp1-v1
    
    aX.append(w0)
    aY.append(w1)

  return aX,aY
```

```python
n=512

x = np.linspace(-2, 2, n)
y = np.linspace(-1, 3, n)

XX, YY = np.meshgrid(x,y)



fig, ax = plt.subplots()

con = ax.contour(XX, YY, f(XX, YY), 10, colors=['black'] ,zorder=10)



ax.clabel(con, inline=True, fontsize=15)


w0,w1 = 0,3
aX, aY = run(w0,w1)
ax.plot(aX, aY, 'o',markersize=2, alpha=0.5, color='red')
print(aX[200],aY[200])

w0,w1 = -1,2
aX, aY = run(w0,w1)
ax.plot(aX, aY, 'o',markersize=1, alpha=0.5, color='yellow')
print(aX[200],aY[200])

w0,w1 = 1,1
aX, aY = run(w0,w1)
ax.plot(aX, aY, 'o',markersize=1, alpha=0.5, color='green')
print(aX[200],aY[200])


w0,w1 = 0,0
aX, aY = run(w0,w1)
ax.plot(aX, aY, 'o',markersize=1, alpha=0.5, color='blue')
print(aX[200],aY[200])

```
