```python
loss_f = lambda x: ùë•**4 - 5*(x**3) + 8*x
derivative = lambda w:(loss_f(w+h)-loss_f(w-h))/(2*h)
```

```python


w=np.linspace(-2,5,1000)


fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot()


w0=8
w1=8

for iter in range(100):
  ax.cla()
  ax.set_title(f'iter={iter}')
  ax.plot(w,loss_f(w))

  h = 0.001
  eta = 0.003
  Beta = 0.5


  w2 = w1-eta*derivative(w1) + Beta*(w1-w0)
  w0 = w1
  w1 = w2
  ax.plot(w2,loss_f(w2),'ro',markersize=8,alpha=0.2)


print(w2)
print(loss_f(w1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML



w = np.linspace(-2,5,1000)
fw = loss_f(w)

fig, ax = plt.subplots()
plt.close()
ax.set_xlim((-2, 5))
ax.set_ylim((-40, 40))

w0=8
w1=8



line1, = ax.plot([], [], lw=2)

dot, = ax.plot([], [], 'ro', ms=10, mec='k')


def init():
    line1.set_data(w, fw)      
    return (line1,)


def animate(i):
  global w0,w1
  h = 0.001
  eta = 0.003
  Beta = 0.8


  w2 = w1-eta*derivative(w1) + Beta*(w1-w0)
  w0 = w1
  w1 = w2

  dot.set_data(w2,loss_f(w2))
  return (dot,)


anim = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=100, blit=True)
rc('animation', html='jshtml')
anim
```
