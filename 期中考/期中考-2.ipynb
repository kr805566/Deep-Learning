```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
def f(x, y):
  return(1-(x/2)+x**5+y**3)*np.exp(-x**2-y**2)
def derivativeX(x, y, h=0.001):
  return (f(x+h,y)-f(x-h,y))/(2*h)
def derivativeY(x, y, h=0.001):
  return (f(x,y+h)-f(x,y-h))/(2*h)
```

```python
def tolerance(x, x0, t=0.5):
  return np.abs(x-x0)<t
```

```python
n = 255
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)

XX, YY = np.meshgrid(x,y)

aX = []
aY = []
bX = []
bY = []

fig, ax = plt.subplots()

ax.set_xlim((-3, 3))
ax.set_ylim((-3, 3))
con = ax.contour(XX, YY, f(XX, YY), 10, colors=['black'] ,zorder=10)
conf = ax.contourf(XX, YY, f(XX, YY), 8, alpha=0.75, cmap=plt.cm.gray)


ax.clabel(con, inline=True, fontsize=15)

dot, = ax.plot([], [], 'ro')

eta = 0.04
beta = 0.968

for ix in x:
  for iy in y:
    w0, w1 = ix, iy
    v0 = w0*eta
    v1 = w1*eta


    for i in range(200):
      wp0, vp0 = w0, v0
      v0 = beta*vp0 + eta*derivativeX(w0, w1)
      wp1, vp1 = w1, v1
      v1 = beta*vp1 + eta*derivativeY(w0, w1)

      w0 = wp0-v0
      w1 = wp1-v1
    
    if tolerance(w0,-1.7) and tolerance(w1,0):
      aX.append(ix)
      aY.append(iy)
    if tolerance(w0,0.1) and tolerance(w1,-1.4):
      bX.append(ix)
      bY.append(iy)

ax.plot(aX,aY,'o',markersize=2, alpha=1, color='red')
ax.plot(bX,bY,'o',markersize=2, alpha=1, color='blue')



plt.show()
```
