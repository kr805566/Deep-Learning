```python
import numpy as np
```

```python
def calc(x, w, b):
   return sum(np.multiply(x,w)+b)
```

```python
def ReLU(z):
  return max(0,z)

def sigmoid(z):
  return 1/(1+np.exp(z))

def softmax(z):
  return np.exp(z)/sum(np.exp(z))
```

```python
x1 = 1
x2 = 3

a1 = sigmoid(calc([x1, x2], [0.1, 0.7], 1.2))
a2 = sigmoid(calc([x1, x2], [0.2, 0.2], 1.5))
a3 = sigmoid(calc([x1, x2], [0.6, 0.4], 0.5))

a4 = ReLU(calc([a1, a2 ,a3], [0.1, 0.6, 0.1], 1.0))
a5 = ReLU(calc([a1, a2 ,a3], [0.4, 0.2, 0.7], -1.0))

y1 = calc([a4, a5], [0.2, 0.5], 0.5)
y2 = calc([a4, a5], [0.3, 0.1], 1.0)
y3 = calc([a4, a5], [0.2, 0.7], 1.2)


Y=softmax([y1,y2,y3])

print("y1={} \ny2={} \ny3={}".format(Y[0], Y[1] ,Y[2]))
```
