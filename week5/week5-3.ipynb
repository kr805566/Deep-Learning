```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
loss_f = lambda x,y: (1-(x/2)+x**5+y**3)*np.exp(-x**2-y**2)

def grad_descent_1D(loss_func, x0, y0, eta, n_iter,h=0.001):

  derivativeX = lambda w:(loss_func(x0+h,y0)-loss_func(x0-h,y0))/(2*h)
  derivativeY = lambda w:(loss_func(x0,y0+h)-loss_func(x0,y0-h))/(2*h)

  weightX = np.zeros(n_iter)
  weightY = np.zeros(n_iter)
  
  lossX = np.zeros(n_iter)
  weightX[0]=x0
  weightY[0]=y0

  loss[0]=loss_func(x0,y0)

  for i in range(1,n_iter):
    x1 = x0-eta*derivativeX(x0)
    y1 = y0-eta*derivativeY(y0)
    weightX[i] = x1
    weightY[i] = y1
    loss[i] = loss_func(x1,y1)
    x0 = x1
    y0 = y1
  print("x={:.5f} y={:.5f} å€¼={:.5f}".format(x0, y0 ,loss_func(x0,y0)))
  return weightX,weightY,loss
```

```python
weightX,weightY = grad_descent_1D(loss_f, -2, -2, 0.5, 100)

x = np.linspace(-3,3,1000)
y = np.linspace(-3,3,1000)

X, Y = np.meshgrid(x,y)
Z = loss_f(X,Y)



fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(1,2,2)

cs = ax.contour(X, Y, Z,[-0.6,-0.4,-0.2,0.0,0.2,0.4,0.6,0.8])

plt.plot(weightX,weightY,'o',markersize=8, alpha=0.2, color='red')
ax.clabel(cs, inline=1, fontsize=8)

ax.set_aspect('equal')
ax.set_title('loss function: f(x,y) = $(1-x/2+x^5+y^3)e^{-x^2-y^2}$')

plt.show()
```
