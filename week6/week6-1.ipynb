```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.normal(1,2,10000)
bins = np.linspace(-8,8,50)
_ = plt.hist(data, bins=bins, edgecolor='w',color='y',alpha=0.5)

data = np.random.normal(0,1,10000)
bins = np.linspace(-8,8,50)
_ = plt.hist(data, bins=bins, edgecolor='w',color='b',alpha=0.5)
```
