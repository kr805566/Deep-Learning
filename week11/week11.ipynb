```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
from tensorflow.keras.utils import plot_model
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np

from keras.layers import Conv2D  # Convolution Operation
from keras.layers import MaxPooling2D # Pooling
from keras.layers import AvgPool2D # Pooling
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout


%matplotlib inline
```

```python
!pip install tflearn
import tflearn.datasets.oxflower17 as oxflower17
X, Y = oxflower17.load_data(one_hot=True)
```

```python
x_train, y_train = X[:1000,:,:,:], Y[:1000,:]
x_test, y_test = X[1000:,:,:,:], Y[1000:,:]

n_classes = 17
```

```python
modelAlexNet = Sequential()  



modelAlexNet.add(Conv2D(96, kernel_size=(11,11), strides=(4,4), activation='relu', padding='valid', input_shape=(224, 224, 3)))
modelAlexNet.add(MaxPooling2D(pool_size=(3,3), strides=(2,2),padding='valid'))


modelAlexNet.add(Conv2D(256, kernel_size=(5,5), padding='same', activation='relu'))
modelAlexNet.add(MaxPooling2D(pool_size=(3,3), strides=(2,2),padding='valid'))


modelAlexNet.add(Conv2D(384, kernel_size=(3,3), padding='same', activation='relu'))
modelAlexNet.add(Conv2D(384, kernel_size=(3,3), padding='same', activation='relu'))
modelAlexNet.add(Conv2D(256, kernel_size=(3,3), padding='same', activation='relu'))
modelAlexNet.add(MaxPooling2D(pool_size=(3,3), strides=(2,2),padding='valid'))


modelAlexNet.add(Flatten())
modelAlexNet.add(Dense(4096, activation='tanh'))
modelAlexNet.add(Dropout(0.5))
modelAlexNet.add(Dense(4096, activation='tanh'))

modelAlexNet.add(Dropout(0.5))

modelAlexNet.add(Dense(n_classes, activation='softmax'))
```

```python
modelAlexNet.compile(loss='categorical_crossentropy', 
       optimizer='adam',
       metrics=['accuracy'])
modelAlexNet.summary()
```

```python
modelAlexNet.fit(x_train, y_train, batch_size=128, epochs=10, verbose=1, validation_data=(x_test, y_test))
```

```python
from keras.layers import BatchNormalization

model104 = Sequential()  

model104.add(Conv2D(96, kernel_size=(11,11), strides=(4,4), activation='relu', input_shape=(224, 224, 3)))
model104.add(MaxPooling2D(pool_size=(3,3), strides=(2,2)))
model104.add(BatchNormalization())

model104.add(Conv2D(256, kernel_size=(5,5), padding='same', activation='relu'))
model104.add(MaxPooling2D(pool_size=(3,3), strides=(2,2)))
model104.add(BatchNormalization())

model104.add(Conv2D(256, kernel_size=(3,3), padding='same', activation='relu'))
model104.add(Conv2D(384, kernel_size=(3,3), padding='same', activation='relu'))
model104.add(Conv2D(384, kernel_size=(3,3), padding='same', activation='relu'))
model104.add(MaxPooling2D(pool_size=(3,3), strides=(2,2)))
model104.add(BatchNormalization())


model104.add(Flatten())
model104.add(Dense(4096, activation='tanh'))
model104.add(Dropout(0.5))
model104.add(Dense(4096, activation='tanh'))

model104.add(Dropout(0.5))

model104.add(Dense(n_classes, activation='softmax'))
```

```python
model104.compile(loss='categorical_crossentropy', 
       optimizer='adam',
       metrics=['accuracy'])
model104.summary()
```

```python
model104.fit(x_train, y_train, batch_size=128, epochs=10, verbose=1, validation_data=(x_test, y_test))
```
