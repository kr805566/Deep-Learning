```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x:3.5*x*(1-x)
```

```python
x = np.linspace(0,1,1000)
f2 = x
n = [1,2,3,6]
sub = 0
strf = 'x'


plt.figure(figsize=(12, 8))

for i in range(6):
  f2 = f(f2)
  strf = 'f({})'.format(strf)

  if (i+1) in n:
    sub+=1
    plt.subplot(220+sub)
    plt.plot(x,f2)
    plt.xlabel('y='+strf)
plt.show()


```
