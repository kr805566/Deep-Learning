```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
from tensorflow.keras.utils import plot_model
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np

from keras.layers import Conv2D  # Convolution Operation
from keras.layers import MaxPooling2D # Pooling
from keras.layers import AvgPool2D # Pooling
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout

%matplotlib inline
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(60000, 28, 28).astype('float32')
x_test = x_test.reshape(10000, 28, 28).astype('float32')


x_train=x_train.reshape(60000,28,28,1).astype('float32')
x_test=x_test.reshape(x_test.shape[0],28,28,1).astype('float32')

x_train /= 255
x_test /= 255

n_classes = 10
y_train = to_categorical(y_train, n_classes)
y_test = to_categorical(y_test, n_classes)
```

```python
modelLeNet = Sequential()  



modelLeNet.add(Conv2D(6, kernel_size=(5,5), activation='sigmoid', padding='valid', input_shape=(28, 28, 1)))
modelLeNet.add(AvgPool2D(pool_size=(2,2), strides=(2,2)))

modelLeNet.add(Conv2D(16, kernel_size=(5,5), activation='sigmoid', padding='valid'))
modelLeNet.add(AvgPool2D(pool_size=(2,2), strides=(2,2)))

modelLeNet.add(Flatten())

modelLeNet.add(Dense(120, activation='sigmoid'))
modelLeNet.add(Dense(84, activation='sigmoid'))

modelLeNet.add(Dense(n_classes, activation='softmax'))
```

```python
modelLeNet.compile(loss='categorical_crossentropy', 
       optimizer=optimizers.SGD(learning_rate=0.01),
       metrics=['accuracy'])
modelLeNet.summary()
```

```python
modelLeNet.fit(x_train, y_train, batch_size=128, epochs=10, verbose=1, validation_data=(x_test, y_test))
```

```python
model103 = Sequential()

model103.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='valid', input_shape=(28,28,1)))
model103.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='valid'))


model103.add(MaxPooling2D(pool_size=(2,2)))
model103.add(Dropout(0.25))
model103.add(Flatten())

model103.add(Dense(128,activation='relu'))
model103.add(Dropout(0.5))

model103.add(Dense(10,activation='softmax'))
```

```python
model103.compile(loss='categorical_crossentropy', 
       optimizer='adam',
       metrics=['accuracy'])
model103.summary()
```

```python
model103.fit(x_train, y_train, batch_size=128, epochs=10, verbose=1, validation_data=(x_test, y_test))
```
