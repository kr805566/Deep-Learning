```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 0.8, 1.6, 2.4, 3.2, 4, 4.8, 5.6, 6.4, 7.2, 8])
y = np.array([1, 9, -2, -31, -38, -45, -55, -38, -25, 16, 71])
```

```python
for i in range(1,11):
  p = np.polyfit(x, y, i)


  xf = np.linspace(0,8,100)
  yf = np.polyval(p,xf)

  plt.plot(x,y,'ro',xf,yf)
  plt.xlabel('{} polynomial'.format(i))
  plt.show()
```
