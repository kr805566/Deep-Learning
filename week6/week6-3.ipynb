```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda z,c:z**2+c
```

```python
n = 513
x = np.linspace(-2,1,n)
y = np.linspace(-2,2,n)

for xx in x:
  for yy in y:
    z0 = 0
    c = xx + yy*1j
    n = 0

    while  (z0<2) and (n<256):
      z0 = f(z0,c)
      n = n+1
    
      
    if (z0<2):
      plt.plot(xx,yy,'o',markersize=1, color='red')

  
plt.show()
```

```python
n = 513
x = np.linspace(-2,1,n)
y = np.linspace(-2,2,n)

X,Y = np.meshgrid(x,y)

c = np.array(X+Y*1j)
z = np.zeros(c.shape,dtype=complex)
b = np.zeros(c.shape,dtype=bool)
m = np.zeros(c.shape)


for _ in range(256):
  z[b] = f(z[b],c[b])
  b = np.abs(z)<2
  m += np.logical_not(b)

plt.figure(figsize=(16,10))
plt.imshow(m,extent=[-2,0.5,-1,1])
plt.show()
```
