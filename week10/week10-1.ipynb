```python
import numpy as np
```

```python
R = [[0.2,0.4,0.3],[0.3,0.9,0.6],[0.9,0.1,0.2]]
Wr = [[0.1,0.2,0.6],[0.6,0.8,0.7],[0.5,0.4,0.3]]

G = [[0.6,0.6,0.9],[0.4,0.7,0.4],[0.7,0.5,0.3]]
Wg = [[0.6,0.6,0.9],[0.4,0.7,0.4],[0.7,0.5,0.3]]

B = [[0.8,0.4,0.9],[0.3,0.1,0.6],[0.8,0.6,0.4]]
Wb = [[0.8,0.4,0.9],[0.3,0.1,0.6],[0.8,0.6,0.4]]
```

```python
R1 = np.pad(R,(1,1))
G1 = np.pad(G,(1,1))
B1 = np.pad(B,(1,1))

Ans=[]
for i in range(3):
  An=[]
  for j in range(3):
    R2 = R1[i:i+3,j:j+3]
    G2 = G1[i:i+3,j:j+3]
    B2 = B1[i:i+3,j:j+3]

    R3 = np.multiply(R2, Wr)
    G3 = np.multiply(G2, Wg)
    B3 = np.multiply(B2, Wb)

    S = np.sum(R3+G3+B3)+0.2
    An.append(S)
  Ans.append(An)

np.around(Ans, 2) 
```
