```python
def oneHotEncode(str):
  strSplits = str.split()
  strLabels = []
  
  for strSplit in strSplits:
    if strSplit not in strLabels:
      strLabels.append(strSplit)

  lenSS=len(strSplits)
  lenSL=len(strLabels)

  arr=[[0]*lenSS for i in range(lenSL)]
  

  for i in range(lenSL):
    for j in range(lenSS):
      if strSplits[j] == strLabels[i]:
        arr[i][j] = 1
  
  print(arr)
 

oneHotEncode('to be or not to be')
```
